```python
#step 1 : Import 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Step 2 : Load and describe dataset
# Dataset used - Academic Performance dataset wherein we monitor how different factors like alcohol consumption affect student performance
# Dataset have 395 rows and 33 columns
df = pd.read_csv('student_mat.csv')
```

```python
#Step 3 : Describe
df.describe()
```

```python
df.columns.tolist()
```

```python
#Step 4 : Missing values
miss = df.isnull().sum()
print(miss[miss > 0])
```

```python
#Step 5 : Check dimensions
print(df.shape[0], df.shape[1])
```

```python
#Step 6 : Check dtypes
df.dtypes
```

```python
#Convert cat to num
from sklearn.preprocessing import LabelEncoder
cat_cols = df.select_dtypes(include='object').columns.tolist()
label_encoder = LabelEncoder()

# Apply label encoding for each categorical column
for col in cat_cols:
    df[col] = label_encoder.fit_transform(df[col])
    
```

```python
df.dtypes
```

```python
bool_cols = df.select_dtypes(include = 'bool').columns.tolist()
df[bool_cols] = df[bool_cols].astype(int)
df.dtypes
```

```python
#Scale numeric cols
from sklearn.preprocessing import MinMaxScaler
num_cols = df.select_dtypes(include = 'number').columns
scaler = MinMaxScaler()
df_scaled = df.copy()
df_scaled[num_cols] = scaler.fit_transform(df[num_cols])
df_scaled.head()
```
