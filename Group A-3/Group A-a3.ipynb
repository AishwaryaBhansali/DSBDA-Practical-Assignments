```python
import pandas as pd
import numpy as np
```

```python
#Load any open source dataset
dataset = pd.read_csv('ageIncome.csv')
```

```python
#Display dataset
df = dataset
df
```

```python
#   1. Provide summary statistics (mean, median, minimum, maximum, standard deviation) for dataset)(age, income, etc) with numeric variables grouped by one of the qualitative variable.
#   For eg, if your categorical variable is age groups and quantitative variable is income,
#   then provide summary statistics of income grouped by the age groups.

#   seperated categorical value and quantitative value
df.loc[df['Age'] < 25, 'Group'] = 'Teenager'
df.loc[df['Age'] >= 25, 'Group'] = 'Adult'
df.loc[df['Age'] >= 65, 'Group'] = 'Senior'

teen = df['Group']== 'Teenager'
adult = df['Group']== 'Adult'
senior = df['Group']== 'Senior'

#Summary statistics of Income of teenagers grouped by age
print("Income earned by Teenagers")
df[teen].describe()
```

```python
#Summary statistics of income of adults grouped by age
print("Income earned by Adults")
df[adult].describe()
```

```python
#Summary statistics of income of seniors grouped by age
print("Income earned by Seniors")
df[senior].describe()
```

```python
#   2. Write a Python program to display some basic statistical details like percentile, mean, standard deviation, etc.
#   of the species of 'Iris-setosa', 'Iris-verginica', 'Iris-versicolor' of iris.csv dataset. 

#load iris dataset
iris = pd.read_csv("iris.csv")
iris
```

```python
#Filtering is done using Boolean indexing.
#Example: iris[iris['species'] == 'setosa'] creates a subset of rows where the species column has the value setosa.
#This technique helps isolate specific data (e.g., Iris-setosa, Iris-versicolor, Iris-virginica) for focused analysis.

#Helps find rows for each flower type (setosa, virginica, versicolor) in the data.
setosa = iris['species'] == 'setosa'
virginica = iris['species'] == 'virginica'
versicolor = iris['species'] == 'versicolor'

#Irirsa setosa species statistical details
print("Iris-setosa")
iris[setosa].head(10) #displaying first 10 records
```

```python
#Iris virginica species statistical details
print("Iris-virginica")
iris[virginica].head(10) #Displaying first 10 records 
```
